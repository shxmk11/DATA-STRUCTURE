**1. 해싱을 할 때 충돌이 발생하면 크게 두 방법으로 해결할 수 있다고 배웠다. 두 방법의 장단점을 서로 비교하여 쓰라.**

    - CHAINING
        장점: 이론적 성능이 좋음
        단점: 연결 리스트라는 별도의 자료 구조 사용

    - OPEN ADDRESSING
        장점: 별도의 자료 구조를 사용하지 않아 간편함
        단점: n차 군집에 취약함

**01 해시 테이블 크기를 1,204로 잡고 해시 함수로 h(x) = x%1024를 사용하였다. 어떤 결점이 있는가?**

    ▶️ 해시 함수값이 0 ~ 1023으로만 나오기 때문에 해시 테이블의 인덱스 1024 ~ 1204가 비게 됨
    ➡️ 해시 테이블 공간 낭비

**02 한 학생이 “다음과 같이 해시 함수를 만들면 확률적으로 고른 분포를 보이는 좋은 함수가 되지 않을까요?”라고 물었다. 단, random(0,10)은 0보다 크고 10 이하인 임의의 실수를 생성한다. h(x) = x*random(0,10)%m 이것은 해시 함수가 될 수 없다. 이유를 설명하시오.**

    ▶️ 해시 함수는 같은 키에 대해 항상 같은 해시값을 반환해야 함
    ▶️ random(1,10)은 매번 다른 값을 생성
    ➡️ 해시 함수의 기본 조건을 만족하지 못하기 때문에 해시 함수가 될 수 없음

**07 체이닝을 사용하는 충돌 해결 방법에서 각 연결 리스트에 새로운 키가 삽입될 때 리스트의 맨 앞에 삽입하지 않고 맨 뒤에 삽입한다면 효율성 면에서 어떤 변화가 생기는가?**

    ▶️ 맨 앞에 삽입시 삽입 시간: O(1)
    ▶️ 맨 뒤에 삽입시 삽입 시간: O(n)
    ➡️ 맨 뒤에 삽입하려면 리스트 끝까지 이동해야 하여 삽입 시간이 더 오래 걸리고 효율성이 떨어진다

**08 체이닝을 사용하는 충돌 해결 방법에서 각 연결 리스트를 정렬된 형태로 유지한다면 효율성 면에서 어떤 변화가 생기는가? (삽입, 삭제, 성공적인 검색, 실패하는 검색의 경우로 나누어 작성하시오.)**

    - 삽입: 정렬 위치를 찾아야 하므로 삽입 성능이 떨어짐
    - 삭제: 탐색 중 조기 종료가 가능해져 평균 삭제 성능이 향상될 수 있음
    - 성공적인 검색: 정렬되어 있어도 원하는 값을 찾을 때까지 탐색해야 해서 큰 차이 x
    - 실패하는 검색: 현재 값이 찾는 값보다 커지는 순간 탐색을 종료할 수 있게 되어 성능 향상
