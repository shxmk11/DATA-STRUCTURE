**Q. 모든 정점 사이에 간선이 존재하는 그래프가 있다. 여러분이라면 인접 행렬 표현과 인접 리스트 표현 중 어느 것을 사용하겠는가? 그리고 그렇게 선택한 이유는 무엇인가?**      

A. 인접 행렬 표현 사용

➡️ 인접 행렬을 사용하면 정점 사이에 간선이 존재하는지 여부를 시간 복잡도로 바로 확인할 수 있기 때문에 훨씬 효율적.

**Q. 정점들 사이에 아주 드물게 간선이 존재하는 그래프가 있다. 이런 그래프를 희소 그래프(Sparse Graph)라 한다. 인접 행렬 표현과 인접 리스트 표현 중 하나를 사용해야 한다면 여러분은 어느 것을 사용하겠는가? 그리고 그렇게 선택한 이유는 무엇인가?**

A. 인접 리스트 표현 사용

➡️ 인접 리스트는 실제로 존재하는 간선 정보만 연결 리스트로 저장하기 때문에 공간 효율성이 극대화.

**Q. 음의 가중치가 있는 그래프에서 다익스트라 알고리즘이 작동하지 않는 예를 들되, 단 3개의 정점을 갖는 그래프를 예로 보이시오.**

A. 
    정점: A, B, C

    간선 및 가중치: 
        A → B (가중치: 2)
        A → C (가중치: 3)
        C → B (가중치: -2)
        
    작동 않는 이유:
        A 방문 처리: B: 2, C: 3 으로 인접한 거리 갱신
        B 방문 처리: 미방문 정점 중 거리값이 가장 작은 정점인 B까지의 최단 거리를 2로 확정되었다고 판단
        C 방문 처리: A → C → B 경로의 거리는 3 + (-2) = 1
        ➡️ A에서 B로 가는 최단 거리는 1인데 이미 B의 거리를 2로 최종 확정했기 때문에 잘못된 결과 도출
