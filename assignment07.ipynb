##### 01 이진 검색 트리에서 균형이 맞지 않으면, 최악의 경우 점근적 검색 시간과 최선의 경우 점근적 검색 시간은 어떻게 되는가?

    1. 최악의 경우
    ▶️ 트리가 한쪽으로 쏠린 경우
    ➡️ O(n)
    2. 최선의 경우
    ▶️ 균형이 맞지 않더라도 완전 이진 트리에 가까운 형태인 경우
    ➡️ O(logn)

##### 02 AVL 트리의 균형은 삽입 또는 삭제 직후에 깨질 수 있다. 다음은 AVL 트리의 균형이 깨진 예다. 그림처럼 균형이 깨진다면 삽입과 삭제 중 어느 한쪽의 작업을 한 후인가, 아니면 둘 다 가능한가?

    ▶️ 노드 70 삽입 or 40의 왼쪽 서브트리에서 노드 삭제 작업으로 인해 균형이 깨졌을 수 있음
    ➡️ 둘 다 가능

##### 03 다음과 같이 AVL 트리의 균형이 깨진 상태라면, 삽입과 삭제 중 어느 한쪽의 작업을 한 후인가, 아니면 둘 다 가능한가?

    ▶️ 노드 17 삽입 or 55의 오른쪽 자식 노드 삭제 작업으로 인해 균형이 깨졌을 수 있음
    ➡️ 둘 다 가능

##### 04 3번 문제의 트리가 AVL 트리가 되도록 수선하고, 수선 과정을 모두 보여라.

    ▶️ 10을 중심으로 좌회전: 20이 10의 부모가 되고 17은 10의 오른쪽 자식이 됨
    ▶️ 40을 중심으로 우회전: 40과 20의 위치를 바꿔 균형 완성
    ➡️ 결과: 20이 루트, 10과 40이 자식, 5, 17, 55이 손자

##### 05 AVL 트리는 균형이 깨진 스타일을 LL, LR, RR, RL의 네 가지 케이스 중 하나로 처리한다. 이들 중 한 번의 회전으로 균형이 맞추어지지 않는 경우는 어느 것인가?

    ▶️ LR, RL은 두 번의 회전이 필요함
    ➡️ LR, RL

##### 06 다음은 맞는 주장인가? - AVL 트리에서 균형이 맞지 않아 수선하고 난 후 수선 전에 비해 AVL 트리의 높이가 반드시 낮아진다.

    ▶️ 트리의 높이가 유지되는 경우도 존재하기 때문에 반드시 낮아지는 것은 아님
    ➡️ 틀린 주장

##### 07 네 가지 타입에 맞추어 AVL 트리를 수선한 후 여전히 균형이 맞지 않아 부모 노드에 대해 수선해줘야 하는 경우가 있다. 이렇게 위로 올라가면서 3회 이상 수선이 필요한 경우가 발생할 수 있는가?

    ▶️ 수선 후에도 해당 서브트리의 전체 높이가 줄어들어 부모 노드에서 다시 불균형이 발생할 수 있기 때문에 3회 의상의 수선이 필요한 경우가 발생할 수도 있음
    ➡️ 발생 가능

##### 08 다음 RB 트리에서 키 7이 삽입되고 균형을 맞춘 후의 모양과 노드의 색상을 그리시오. 편의상 RB트리에서 사용하는 블랙 리프들은 그리지 않았다.



##### 09 RB트리에서 리프 노드에 새 키를 삽입하고 레드를 칠하니 성질3이 깨졌다. 이를 수선하는 과정에서 최악의 경우에는 루트까지 영향을 미칠 수도 있다. 루트까지 영향을 미치는 상황을 설명해 보시오. 삽입이 일어난 노드를 p라 할 때 적어도 p6까지는 어떤 일이 일어나는지 보여야 한다. 여기서 p6은 p에서 루트로 가는 경로에서 다섯 번째 만나는 조상 노드, 즉 p의 고고조할머니가 된다.

    ▶️ p = RED, p1 = RED, p1의 형제 = RED
    ▶️ p1 → BLACK, p1의 형제 → BLACK, p2 → RED
    ▶️ p2 = RED, p3 = RED, p3의 형제 = RED
    ▶️ p3 → BLACK, p3의 형제 → BLACK, p4 → RED
    ▶️ p4 = RED, p5 = RED, p5의 형제 = RED
    ▶️ p5 → BLACK, p5의 형제 → BLACK, p6 → RED
    ▶️ p6 = RED, p7 = RED, p7의 형제 = RED
    ➡️ RED-RED 위반이 계속 한 단계 위로 이동
    ➡️ 최악의 경우 계속 반복해서 루트까지 올라가서 영향을 미침

##### 15 디스크의 한 블록이 8,192바이트, 키의 크기가 20바이트, 페이지 번호가 4바이트를 차지한다면, B-트리의 한 노드는 최대 몇 개의 키를 가질 수 있도록 설계하겠는가?

    ▶️ 20k + 4(k + 1) ≤ 8192
    ▶️ 24k ≤ 8188
    ▶️ k ≤ 341
    ➡️ 최대 341개의 키

##### 16 각 노드가 최대 k개의 키를 가질 수 있는 B-트리에서 리프 노드에 들어 있는 키의 수와 내부 노드에 들어 있는 키의 수는 어떤 관계가 있는가?

    ▶️ 분할 과정에서 올라온 일부 키만 내부 노드에 들어가기 때문에 대부분의 키가 리프 노드에 들어가게 됨
    ➡️ 내부 노드에 들어 있는 키의 수는 리프 노드에 들어 있는 키의 수보다 현저히 적음

##### 17 B-트리에서 언더플로우가 일어나 자신의 형제 노드와 병합할 때 두 노드의 키들을 모두 합치고 부모 노드에서 추가로 1개의 키를 가져온다. 이렇게 다 합치면 병합으로 만들어진 노드에서 오버플로우가 생기는 경우는 없는가?

    ▶️ 두 노드와 부모 노드의 키 1개를 전부 합쳐도 최대 키를 넘지 않도록 설계됨

##### 18 B-트리의 각 노드가 담을 수 있는 최대 키 수가 1,000개인 예에서 1조 개의 키로 덱싱을 하면 B-트리의 높이의 범위는 어떻게 되는가? 단, 하나의 노드로 이루어진 트리의 높이를 1로 정의한다.

    ▶️ 1조 = 10¹²
    ▶️ 최소 높이: log₁₀₀₀(10¹²) = log(10¹²)/log(10³) = 12/3 = 4
    ▶️ 최고 높이: log₅₀₁(10¹²) = log(10¹²) / log(501) = 12 / 2.7 ≈ 4.4 → 5(올림)
    ➡️ 높이 범위: 4 ~ 5
